```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

# Function to prepare data for Prophet
def prepare_data_for_prophet(data, column):
    prophet_df = pd.DataFrame({
        'ds': pd.to_datetime(data['datum']),
        'y': data[column]
    }).dropna()  # Drop any rows where the target column has NaN values
    return prophet_df

# Function to shift dates to the end of the month
def shift_to_month_end(dates):
    return dates + pd.offsets.MonthEnd(0)

# Function to plot original time series
def plot_time_series(df, medicine_column):
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='datum', y=medicine_column)
    plt.title(f'Time Series Plot for {medicine_column}')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to train Prophet model
def train_prophet_model(data, column, periods):
    prophet_df = prepare_data_for_prophet(data, column)
    
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        seasonality_mode='multiplicative',
        interval_width=0.95
    )
    
    model.add_seasonality(name='monthly', period=30.5, fourier_order=5)
    
    model.fit(prophet_df)
    
    future_dates = model.make_future_dataframe(periods=periods, freq='M')  # Generate future dates
    
    forecast = model.predict(future_dates)
    
    return model, forecast

# Function to plot results
def plot_results(model, forecast, actual_data, column, save_path=None):
    fig = model.plot(forecast)
    plt.title(f'Prophet Forecast for {column}')
    plt.plot(actual_data['ds'], actual_data['y'], 'r.', alpha=0.5, label='Actual')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}/{column}_forecast.png')
        plt.close()
    else:
        plt.show()

    fig_comp = model.plot_components(forecast)
    plt.tight_layout()
    plt.show()

# Function to evaluate model performance
def evaluate_model(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return mae, rmse, mape

# Main function with diagnostics
def main():
    medicine_columns = [col for col in df.columns if col != 'datum']
    results = {}
    
    for medicine in medicine_columns:
        print(f"\nProcessing {medicine}...")

        try:
            plot_time_series(df, medicine)
            
            # Split train and test data
            train_size = int(len(df) * 0.8)
            train_df = df.iloc[:train_size]
            test_df = df.iloc[train_size:]
            
            # Calculate the number of months for forecasting (based on test set length)
            periods = len(test_df)  # Ensure you forecast the exact number of months in the test set
            
            # Train the model
            model, forecast = train_prophet_model(train_df, medicine, periods=periods)
            
            # Prepare test data and forecast
            test_prophet_df = prepare_data_for_prophet(test_df, medicine)
            test_dates = shift_to_month_end(test_prophet_df['ds'])  # Shift test dates to month-end
            
            # Debugging: Print test dates and forecast dates
            print(f"Test dates for {medicine} (after shifting to month-end): {test_dates}")
            print(f"Future forecast dates for {medicine}: {forecast['ds'].tail(periods)}")
            
            # Filter forecast for the test period
            forecast_test = forecast[forecast['ds'].isin(test_dates)]
            
            # Ensure matching lengths by trimming forecast or test set
            min_length = min(len(test_prophet_df), len(forecast_test))
            test_prophet_df = test_prophet_df.iloc[:min_length]
            forecast_test = forecast_test.iloc[:min_length]

            # Check if forecast_test has valid entries
            if not forecast_test.empty:
                mae, rmse, mape = evaluate_model(
                    test_prophet_df['y'].values,
                    forecast_test['yhat'].values
                )
                
                results[medicine] = {
                    'mae': mae,
                    'rmse': rmse,
                    'mape': mape,
                    'model': model,
                    'forecast': forecast
                }

                # Plot results
                prophet_df = prepare_data_for_prophet(df, medicine)
                plot_results(model, forecast, prophet_df, medicine)
                
                print(f"Results for {medicine}:")
                print(f"MAE: {mae:.2f}")
                print(f"RMSE: {rmse:.2f}")
                print(f"MAPE: {mape:.2f}%")
            else:
                print(f"No valid predictions for {medicine}. Check data alignment.")
            
        except Exception as e:
            print(f"Error processing {medicine}: {str(e)}")
    
    return results

if __name__ == "__main__":
    results = main()
```
